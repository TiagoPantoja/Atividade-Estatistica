```python
import pandas as pd


data = {
    'NumeroErros': [0, 1, 2, 3, 4],
    'Frequencia': [25, 20, 3, 1, 1]
}

df = pd.DataFrame(data)

media_erros = (df['NumeroErros'] * df['Frequencia']).sum() / df['Frequencia'].sum()

print("Questão a) Número médio de erros por página:", media_erros)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"Erros": [0, 1, 2, 3, 4], "Frequência": [25, 20, 3, 1, 1]})

mediana = df["Erros"].median()

print("Questão b) Número mediano de erros:",  mediana)
```

```python
import pandas as pd
import numpy as np

data = {
    'NumeroErros': [0, 1, 2, 3, 4],
    'Frequencia': [25, 20, 3, 1, 1]
}

df = pd.DataFrame(data)

valores = []
for idx, row in df.iterrows():
    valores.extend([row['NumeroErros']] * row['Frequencia'])

desvio_padrao = np.std(valores)

print(" Questão c) Desvio padrão dos erros por página:", desvio_padrao)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'NumeroErros': [0, 1, 2, 3, 4],
    'Frequencia': [25, 20, 3, 1, 1]
}

df = pd.DataFrame(data)

valores = []
for idx, row in df.iterrows():
    valores.extend([row['NumeroErros']] * row['Frequencia'])

plt.hist(valores, bins=range(6), color = 'green', edgecolor='white')
plt.xlabel('Número de Erros')
plt.ylabel('Frequência')
plt.title('Questão d) Gráfico do Número de Erros por Página')
plt.xticks(range(5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

num_casas = [2, 2, 3, 10, 13, 14, 15, 15, 16, 16, 18, 18, 20, 21, 22, 22, 23, 24, 25, 25, 26, 27, 29, 29, 30, 32, 36, 42, 44, 45, 45, 46, 48, 52, 58, 59, 61, 61, 61, 65, 66, 66, 68, 75, 78, 80, 89, 90, 92, 97]

num_bins = 5
hist, bins = np.histogram(num_casas, bins=num_bins)
relative_freq = hist / len(num_casas)

interval_labels = [f'{bins[i]} - {bins[i+1]}' for i in range(len(bins)-1)]

bar_width = 1.0
bar_positions = np.arange(len(interval_labels))

plt.bar(bar_positions, relative_freq, color='green', edgecolor='white', width=bar_width)
plt.xlabel('Número de Casas por Quarteirão')
plt.ylabel('Frequência')
plt.title('Questão 2 a) Histograma do Número de Casas por Quarteirão')
plt.xticks(bar_positions, interval_labels, rotation=45)
plt.tight_layout()
plt.show()
```
